```python
import pickle

file = open("pendulumSim", "rb")

solE0 = pickle.load(file)
solE10 = pickle.load(file)
solEP = pickle.load(file)
```

```python
from matplotlib import pyplot as plt
def plot_motion(t,theta1,theta2,p_theta1,p_theta2):
    fig, (theta, p) = plt.subplots(2,1, figsize = (8,8))
    theta.plot(t, theta1, label = r"$\theta_1$")
    theta.plot(t, theta2, label = r"$\theta_2$")
    theta.set_xlabel("time [s]")
    theta.set_ylabel("Radian")
    theta.legend()
    theta.grid()

    p.plot(t, p_theta1, label = r"$p_{\theta_1}$")
    p.plot(t, p_theta2, label = r"$p_{\theta_2}$")
    p.set_xlabel("time [s]")
    p.set_ylabel("angular momentum $[kg m^2 / s]$")
    p.legend()
    p.grid()
```

```python
plot_motion(*solE10)
```

```python
plot_motion(*solEP)
```
